```python
#경력년월수(개월) 변수 만들기
career = ['CAREER_1_YRCNT','CAREER_1_MMCNT','CAREER_2_YRCNT','CAREER_2_MMCNT','CAREER_3_YRCNT','CAREER_3_MMCNT']
df[career] = df[career].fillna(0)
df['CAREER_YMCNT'] = df['CAREER_1_YRCNT']*12 +df['CAREER_1_MMCNT'] + df['CAREER_2_YRCNT']*12 +df['CAREER_2_MMCNT'] +df['CAREER_3_YRCNT']*12 +df['CAREER_3_MMCNT']
```

```python
# 희망지역 코드 두자리 변수 생성
df['HOPE_REGION_1_2CD'] = df['HOPE_REGION_1_CD'].str[0:2]

# 거주지 코드 두자리 변수 생성
df['REGION_2CD'] = df['REGION_CD'].str[0:2]

# 희망지역 - 거주지 일치 여부(일치: True, 불일치: False)
df['REGION_EQUAL'] = (df['HOPE_REGION_1_2CD'] == df['REGION_2CD'])
```

```python
# 고용형태코드 리코딩
# 기간의 정함이 없는 근로계약(10) or 기간의 정함이 없는 근로계약(시간(선택)제)(11) --> 1 (정규직)
# 기간의 정함이 있는 근로계약(20) or 기간의 정함이 있는 근로계약(시간(선택)제)(21) --> 2 (비정규직)
# 관계없음(99) --> 0  

def test1(x):
  if (x == '10') | (x == '11') :
    return 1
  elif (x == '20') | (x == '21') :
    return 2
  elif x == '99':
    return 0

df['RE_EMP_TP_CD'] = df['EMP_TP_CD'].apply(test1)
```

```python
# 희망고용형태코드
print(df['RE_EMP_TP_CD'].value_counts())

# 0: 관계없음 
# 1: 정규직
# 2: 비정규직
```
